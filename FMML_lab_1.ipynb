Question 0:
write down 3-5 methods applicable to each data type. (Hint: extend, reverse, etc.

```python
# Lists
my_list = [11, 22, 33]

# append(): Adds an element to the end of the list.
my_list.append(44)
print("append():", my_list)

# extend(): Appends the elements of an iterable to the end of the list.
my_list.extend([55, 66])
print("extend():", my_list)

# insert(): Inserts an element at a specific index in the list.
my_list.insert(11, 77)
print("insert():", my_list)

# pop(): Removes and returns the element at a specified index.
popped_element = my_list.pop(2)
print("pop():", popped_element, my_list)

# reverse(): Reverses the order of elements in the list.
my_list.reverse()
print("reverse():", my_list)


# Strings
my_string = "Raja Muneesh Kanth"

# lower(): Converts all characters in the string to lowercase.
lowercase_string = my_string.lower()
print("lower():", lowercase_string)

# upper(): Converts all characters in the string to uppercase.
uppercase_string = my_string.upper()
print("upper():", uppercase_string)

# split(): Splits the string into a list of substrings based on a specified delimiter.
split_string = my_string.split(", ")
print("split():", split_string)

# join(): Joins a list of strings into a single string using the string as a separator.
joined_string = "-".join(split_string)
print("join():", joined_string)

# replace(): Replaces occurrences of a substring with another substring.
replaced_string = my_string.replace("Muneesh", "Raja")
print("replace():", replaced_string)


# Dictionaries
my_dict = {'name': 'Raja', 'age': 21, 'city': 'Hyderababd'}

# keys(): Returns a list of all keys in the dictionary.
keys_list = list(my_dict.keys())
print("keys():", keys_list)

# values(): Returns a list of all values in the dictionary.
values_list = list(my_dict.values())
print("values():", values_list)

# items(): Returns a list of key-value pairs as (key, value) tuples.
items_list = list(my_dict.items())
print("items():", items_list)

# get(): Retrieves the value associated with a given key.
age = my_dict.get('age')
print("get():", age)

# update(): Updates the dictionary with key-value pairs from another dictionary.
my_dict.update({'city': 'Mumbai', 'job': 'DSAI'})
print("update():", my_dict)


# Sets
my_set1 = {1, 2, 3}
my_set2 = {3, 4, 5}

# add(): Adds an element to the set.
my_set1.add(6)
print("add():", my_set1)

# remove(): Removes a specified element from the set; raises an error if not present.
my_set1.remove(3)
print("remove():", my_set1)

# discard(): Removes a specified element from the set if present; no error if not found.
my_set2.discard(5)
print("discard():", my_set2)

# union(): Returns a new set containing all unique elements from both sets.
union_set = my_set1.union(my_set2)
print("union():", union_set)

# intersection(): Returns a new set containing elements common to both sets.
intersection_set = my_set1.intersection(my_set2)
print("intersection():", intersection_set)
```

Define a function, which takes in two strings A and B. Reverses the first string A, adds it to B, and returns the final string

```python
def raja(X, Y):
    reversed_X = X[::-1]
    final_string = reversed_X + Y
    return final_string
string_X = "FMML"
string_Y = " LAB1"
result = raja(string_X, string_Y)
print(result)
```

Question 2 :
Given a list having Names, work_hours, and gender, Write a function to print name of the female worker that worked the most hours. Also how much do should she be paid if the pay is $ 20 per hour.
work_hours = [('Abby',100 , 'F'),('Billy',400, 'M'),('Cassie',800,'F'), ('Maggi',600,'F'),('Alex',500,'M'),('Raj',225,'M'),('Penny',920,'F'),('Ben',300,'M')]
Answer : the female worker that worked the most hours is Penny and she should be paid 18400

```python
def raja(work_hours):
    max_hours = 0
    female_worker = None
    for name, hours, gender in work_hours:
        if gender == 'F' and hours > max_hours:
            max_hours = hours
            female_worker = name
    if female_worker:
        payment = max_hours * 20
        return female_worker, payment
    else:
        return "No female workers found in the list", 0
work_hours = [('Abby', 100, 'F'), ('Billy', 400, 'M'), ('Cassie', 800, 'F'), ('Maggi', 600, 'F'), ('Alex', 500, 'M'), ('Raj', 225, 'M'), ('Penny', 920, 'F'), ('Ben', 300, 'M')]

female_worker, payment = raja(work_hours)

print(f"The female worker that worked the most hours is {female_worker} and she should be paid {payment}")
```
